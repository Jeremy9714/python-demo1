## 1. 图表

```python
"""
折线图(plot) 趋势随时间变化
柱状图(bar) 类别之间对比
饼图(pie) 整体组成比例
散点图(scatter) 两变量相关性
箱线图(boxplot) 数据分布、异常
组合图
"""
```

### 1.1 折线图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 运行时配置相关

# rcParams['font.family'] = 'SimHei'  # 字体支持中文
rcParams['font.sans-serif'] = 'SimHei'  # 字体支持中文

# 创建图表
plt.figure(figsize=(10, 5))  # 设置大小(画布宽、高)

# 绘图所需数据
month = ['一月产品销售数据', '二月产品销售数据', '三月产品销售数据', '四月产品销售数据']
sales = [100, 150, 80, 120]

# 绘制折线图
plt.plot(
    month,  # X轴内容
    sales,  # Y轴内容
    label='产品A',  # 图例内容
    color='orange',  # 折线颜色
    linewidth=2,  # 线宽
    linestyle='-',  # 线风格
    marker='o'  # 数据点风格
)

# 添加标题
plt.title('Q1 sales trending', color='red', fontsize=20)

# 添加坐标轴标签
plt.xlabel('月份', fontsize=10)  # x-axis标签
plt.ylabel('销售额(万元)', fontsize=10)  # y-axis标签

# 添加图例
plt.legend(loc='upper left')  # 左上方显示

# 添加网格线
plt.grid(
    True,
    alpha=0.2,  # 透明度
    color='blue',  # 颜色
    linestyle='--'  # 网格线风格 '-' 实线 '--' 虚线
)
# plt.grid(axis='x')  # X轴网格线
# plt.grid(axis='y')  # Y轴网格线

# 设置刻度字体信息
plt.xticks(
    rotation=45,  # 字体角度
    fontsize=10,  # 字体大小
    color='black'  # 字体颜色
)
plt.yticks(
    rotation=0,
    fontsize=10
)

# 设置坐标轴范围
plt.ylim(60, 160)  # Y轴范围

# 设置数据点显示内容
for x, y in zip(month, sales):
    plt.text(
        x,  # X轴位置
        y + 2,  # Y轴位置
        str(y),  # 显示内容
        ha='center',  # 水平对齐
        va='bottom',  # 垂直对齐
        fontsize=10
    )

# 显示图表
plt.show()
```

### 1.2 柱状图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['font.sans-serif'] = 'SimHei'

plt.figure(figsize=(10, 5))

subject = ['数学', '英语', '物理', '化学', '生物']
grade = [100, 99, 85, 90, 95]

plt.bar(subject, grade, label='学生成绩', width=0.6, color='orange', zorder=10)
plt.title('Q1 Mock exam results', fontsize=20, color='red')
plt.xlabel('科目', fontsize=15)
plt.ylabel('分数', fontsize=15)
plt.legend(loc='upper right')
plt.grid(axis='y', alpha=0.2, color='blue', linestyle='--', zorder=0)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylim(0, 110)
for x, y in zip(subject, grade):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=10)

# 自动调整布局
plt.tight_layout()

plt.show()
```

### 1.3 条形图

```python
countries = ['PRC', 'USA', 'UK', 'JPN', 'IND']
gdp = [90, 105, 80, 85, 70]

plt.barh(countries, gdp, label='GDP', height=0.8, color='orange', zorder=10)
plt.title('2024 GDP Ranking', color='red', fontsize=20)
plt.xlabel('美元(万亿)', fontsize=15, color='black')
plt.ylabel('国家', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc='upper right')
plt.grid(axis='x', alpha=0.2, color='black', linestyle='--', zorder=0)
plt.xlim(0, 120)
plt.tight_layout()

plt.show()
```

### 1.4 饼图
### 1.4.1 饼图

```python
todos = ['学习', '吃饭', '娱乐', '运动', '睡觉']
time = [6, 2, 5, 3, 8]
colors = ['red', 'orange', 'yellow', 'green', 'purple']

plt.pie(
    time,
    labels=todos,
    autopct='%.1f%%',  # 显示百分比的格式(%转义符)
    startangle=90,  # 初始图画的角度
    colors=colors,  # 设置颜色
)
plt.title('Daily time distribution of todos', color='red', fontsize=20)
plt.tight_layout()

plt.show()
```

### 1.4.2 环形图

```python
todos = ['学习', '吃饭', '娱乐', '运动', '睡觉']
time = [6, 2, 5, 3, 8]
colors = ['red', 'orange', 'yellow', 'green', 'purple']

plt.pie(
    time,
    labels=todos,
    autopct='%.1f%%',
    colors=colors,
    wedgeprops={
        'width': 0.5  # 设置圆环宽度
    },
    pctdistance=0.7  # 百分比显示位置(距离圆心)
)
plt.title('Daily time distribution of todos', color='blue', fontsize=20)
# 环形图文本注释
plt.text(0, 0, str(sum(time)) + '小时', ha='center', va='center', fontsize=15)
plt.tight_layout()

plt.show()
```

### 1.4.3 爆炸环形图

```python
todos = ['学习', '吃饭', '娱乐', '运动', '睡觉']
time = [6, 2, 5, 3, 8]
colors = ['red', 'orange', 'yellow', 'green', 'purple']
explodes = [0.1, 0, 0, 0, 0]  # 饼块距离圆心的位置

plt.pie(
    time,
    labels=todos,
    autopct='%.1f%%',
    startangle=90,
    colors=colors,
    explode=explodes,  # 爆炸饼图效果
    shadow=True  # 阴影效果
)
plt.title('Daily time distribution of todos', color='blue', fontsize=20)
plt.tight_layout()

plt.show()
```

### 1.5 散点图

```python
import random

# from matplotlib import rcParams

# rcParams['font.sans-serif'] = 'SimHei'

plt.figure(figsize=(10, 8))

x = []
y = []
for i in range(1000):
    tmp = random.uniform(0, 10)
    x.append(tmp)
    y.append(2 * tmp + random.gauss(0, 2))

plt.scatter(
    x,
    y,
    color='blue',
    alpha=0.4,  # 透明度
    s=20,  # 圆点大小
    label='散点图'
)
plt.plot([0, 10], [0, 20], label='折线图', color='red', linewidth=2, linestyle='--')
plt.title('Linear regression scatter', color='orange', fontsize=20)
plt.xlabel('Independent Variable', fontsize=15, color='black')
plt.ylabel('Dependent Variable', fontsize=15, color='black')
plt.xticks(rotation=0, fontsize=15)
plt.yticks(rotation=0, fontsize=15)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.1, color='blue', linestyle='--')

plt.tight_layout()

plt.show()
```

### 1.6 箱线图

```python
data = {
    '语文': [82, 85, 88, 70, 90, 76, 84, 83, 95],
    '数学': [75, 80, 79, 93, 88, 82, 87, 89, 92],
    '英语': [70, 72, 68, 65, 78, 80, 85, 90, 95]
}

plt.figure(figsize=(8, 6))
plt.boxplot(data.values(), labels=data.keys())
plt.title('Grade distribution(boxplot)', color='red', fontsize=20)
plt.ylabel('分数')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

plt.show()
```

### 1.7 组合图

```python
month = ['1', '2', '3', '4']
sales = [100, 150, 80, 120]

"""
subplot(rows,cols,index)
逐个创建子图，通过 rows, cols, index 指定子图在网格中的位置
"""
f1 = plt.subplot(2, 2, 1)  # 第一个子图
f1.plot(month, sales)
f2 = plt.subplot(222)  # 第一个子图
f2.bar(month, sales)
f3 = plt.subplot(223)  # 第一个子图
f3.barh(month, sales)
f4 = plt.subplot(224)  # 第一个子图
f4.scatter(month, sales)
```

```python
# fig, axes = plt.subplots(2, 2, figsize=(10, 6))
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 6))
# axes[0, 0].set_title('s1')
# axes[0, 1].set_title('s2')
# axes[1, 0].set_title('s3')
# axes[1, 1].set_title('s4')
ax1.plot(month, sales)
ax1.set_title('折线图')
ax2.bar(month, sales)
ax2.set_title('柱状图')
ax3.barh(month, sales)
ax3.set_title('条形图')
ax4.scatter(month, sales)
ax4.set_title('散点图')

plt.tight_layout()
plt.show()
```
